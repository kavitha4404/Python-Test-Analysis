### Find out the employee number and dept of employee who does overtime?

```python
# Filter employees who do overtime
overtime_employees = df[df['OverTime'] == 'Yes']

# Select EmployeeNumber and Department columns for overtime employees
result = overtime_employees[['EmployeeNumber', 'Department']]

# Display the result
result
```

### Find out last 5 employees based on last promotion received?

```python
# Sort the dataset by 'YearsSinceLastPromotion' in descending order
sorted_data = df.sort_values(by='YearsSinceLastPromotion',ascending = False)
sorted_data.head(5)
```

### Find out the list of employee whose income is more than the average income of all the employee’s present in the same department?

```python
# calculating the average income for each department
avg_income_by_department = df.groupby('Department')['MonthlyIncome'].mean().reset_index()
avg_income_by_department.rename(columns={'MonthlyIncome':'AverageIncome'},inplace= True)

# Filter employees whose income is more than the average income of their department
filtered_employees = df[df['MonthlyIncome'] > df['AverageIncome']]

# Select relevant columns for the result (e.g., EmployeeNumber, Department, MonthlyIncome)
result = filtered_employees[['EmployeeNumber', 'Department', 'MonthlyIncome']]

# Display the result
print(result)
```

### To find the employee who are the highest salary in monthly income

```python
# To find the employee who are the highest salary in monthly income
highest_salary_employee = df[df['MonthlyIncome']==df['MonthlyIncome'].max()].reset_index()
highest_salary_employee[['EmployeeNumber','Department','Age','Gender','JobRole','MonthlyIncome']]
```

### Scatter Plot of Total Working Years vs. Monthly Income

```python
# Scatter Plot of Total Working Years vs. Monthly Income
sns.scatterplot(x='TotalWorkingYears', y='MonthlyIncome', data=df, alpha=0.5)

# Set labels and title
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.title('Scatter Plot of Total Working Years vs. Monthly Income')
plt.show()
```

### To find Employees, how many employees are worked as “Research Scientist” and “Sales Executive”.

```python
# filter the data employees with 'JobRole' as "Research Scientist" or "Sales Executive"
filtered_data = df[(df['JobRole']=='Research Scientist')|(df['JobRole']=='Sale Executive')]
# count the number of employees
count_employees = len(filtered_data)
print(f"Number of employees who worked as 'Research Scientist' or 'Sales Executive': {count_employees}")
```

### To find the employee marital status as ascending to descending order by using dataset.

```python
ascending_order = df.sort_values(by='MaritalStatus', ascending=True)
descending_order = df.sort_values(by='MaritalStatus', ascending=False)

# Display the ascending order
print("Ascending Order:")
print(ascending_order[['EmployeeNumber', 'MaritalStatus']])

# Display the descending order
print("\nDescending Order:")
print(descending_order[['EmployeeNumber', 'MaritalStatus']])
```

### Two subplots in one figure by using the dataset.

```python
# Subplot 1: Histogram of Age
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=20, color='skyblue',edgecolor='blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Subplot 2: Boxplot of MonthlyIncome
plt.subplot(1, 2, 2)
sns.histplot(df['MonthlyIncome'],bins=20,color='yellow',edgecolor='orange')
plt.xlabel('Monthly Income')
plt.title('Monthly Income Distribution')
# Show the combined figure
plt.show()
```

### Boxplot by using dataset.

```python
# Create a boxplot for MonthlyIncome by JobRole
plt.figure(figsize=(12, 6))
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df, palette='Set3')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.title('Boxplot of Monthly Income by Job Role')
plt.xticks(rotation=45) 
plt.show()
```
