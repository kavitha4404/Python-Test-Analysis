## step3: How many Pokémon are legendary?

```python
# Filter the DataFrame to select legendary Pokemon
legendary_pokemons = df[df["is_legendary"]==1]
# count of legendry pokemons
num_legendary_pokemons = len(legendary_pokemons)
print("The number of legendary pokemons are:",num_legendary_pokemons)
```

## step4: Legendary Pokémon by type?

```python
# filter the dataframe of legendary pokemons
legendary_pokemon = df[df["is_legendary"]==1]
# group the lengendary pokemon by their(type1)
legendary_pokemon_by_type = legendary_pokemon.groupby('type1').size().reset_index(name='count')
# sort the data by decesending order
legendary_pokemon_by_type.sort_values(by='count',ascending=False)
```

```python
# filter the dataframe of legendary pokemons
legendary_pokemon = df[df["is_legendary"]==1]
# group the lengendary pokemon by their(type2)
legendary_pokemon_by_type = legendary_pokemon.groupby('type2').size().reset_index(name='count')
# sort the data by decesending order
legendary_pokemon_by_type.sort_values(by='count',ascending=False)
```

## step5: Legendary Pokémon by fighter stats?

```python
# Filter the DataFrame to select legendary Pokémon
legendary_pokemon = df[df['is_legendary'] == 1]

# Calculate a combined fighting stat by summing the relevant columns
legendary_pokemon['combined_fighting_stat'] = (
    legendary_pokemon['attack'] +
    legendary_pokemon['defense'] +
    legendary_pokemon['sp_attack'] +
    legendary_pokemon['sp_defense'] +
    legendary_pokemon['speed']
)
# Sort the legendary Pokémon by their combined fighting stat
legendary_pokemon = legendary_pokemon.sort_values(by='combined_fighting_stat', ascending=False)

# Display the top legendary Pokémon based on their combined fighting stat
print("Top Legendary Pokémon by Combined Fighting Stat:")

print(legendary_pokemon[['name', 'type1', 'type2', 'combined_fighting_stat']].head(10))
```

## step8: What are the strongest and weakest Pokemon species?

```python
# Find the strongest pokemon species based on base_stats
#(which include  attributes as attack,sp_attack,sp_defense,defense,speed) 
strongest_pokemon = df[df['base_total']==df['base_total'].max()]
# Find the weakest pokemon species based on base_stats
weakest_pokemon = df[df['base_total']==df['base_total'].min()]
print("The strongest pokemon species are:")
print(strongest_pokemon[["name","base_total"]])
print("The weakest pokemon species are:")
print(weakest_pokemon[["name","base_total"]])
```

# OR

```python

# Sort Pokémon by their total base stats
strongest_pokemon = df.nlargest(10, 'base_total')
weakest_pokemon = df.nsmallest(10, 'base_total')

# Concatenate the strongest and weakest Pokémon dataframes
combined_pokemon = pd.concat([strongest_pokemon, weakest_pokemon])

# Create a seaborn bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=combined_pokemon, x='base_total', y='name', hue='base_total', palette='viridis')
plt.title('Top 10 Strongest and Weakest Pokémon Species')
plt.xlabel('Total Base Stats')
plt.ylabel('Pokémon Species')
plt.legend(title='Total Base Stats', loc='lower right')
plt.show()
```

## step9: What are the strongest and weakest types of Pokemon?

```python
# List of columns representing type effectiveness
type_columns =['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water']

# caluclate the average effectiveness of the each type
type_effectiveness=df[type_columns].mean()

# find the strogest and weakest pokemon types
strongest_pokemon=type_effectiveness.idxmax()
weakest_pokemon=type_effectiveness.idxmin()
print("The strongest type :",strongest_pokemon)
print("The weakest type :",weakest_pokemon)
```

```python
type_effectiveness.plot(kind='bar',color='purple')
```

# OR

```python
# group the pokemon by thier primary type(type1) and calculate the average total base stats
by_type_pokemon=df.groupby("type1")["base_total"].mean().reset_index()
# sort the data by base total in decsending order
sort_pokemon=by_type_pokemon.sort_values(by="base_total",ascending=False)
# creating a bar plot
plt.figure(figsize=(12,6))
sns.barplot(data=sort_pokemon,x='type1',y='base_total',palette='viridis')
plt.title("Average Total Base Stats by Pokémon Type")
plt.xlabel("pokemon type")
plt.ylabel("avg_base_total_stats")
plt.show()
```

## step11: Are any of the statistics correlated?

```python

# Select the statistics to analyze
statistics_to_analyze = ['attack', 'defense', 'speed']

# Calculate the correlation matrix
correlation_matrix = df[statistics_to_analyze].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Pokémon Statistics')
plt.show()
```

## step12: Create a training/test split?


To create a training/test split for your Pokémon dataset in Python, you can use libraries such as scikit-learn


```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# defining the features(X) and target variables (y)
# for example let's say you want to predict the 'is_legendary' pokemon statistics based on 'attack', 'defense' and 'speed'
features =['attack','defense','speed']
target ='is_legendary'
X = df[features]
y = df[target]
# Split the dataset into a training set and a test set (e.g., 70% training, 30% testing)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Print the shapes of the resulting sets
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```
