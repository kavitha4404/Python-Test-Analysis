## Find	City wise list all the Villa which is not less than ten thousand

```python
# filter the data to includes only villas
filter_data = df[(df['type'] == 'villa') & (df['price'] >= 10000)]

# grouping the data by each city
grouped_data = filter_data.groupby('city')
for city,city_data in grouped_data:
    print(f"City: {city}")
    for index, row in city_data.iterrows():
        print(f"Street: {row['street']}, Price: {row['price']}, Beds: {row['beds']}, Baths: {row['baths']}, Sq.Ft: {row['sq__ft']}")
    print('\n')
```

## In GALT city which residential type has more than 800sq__ft. Display their respective details street,sq__ft,sale_date,city.



```python
filtered_data = df[(df['city'] == 'GALT') & (df['sq__ft'] > 800)]
results=filtered_data[['street','city','sq__ft','sale_date']]
results
```

## Which is the cheapest Villa in CA. name the city,street and price for the Villa.



```python
# Filter the dataset to include only villas in California (CA)
ca_villas = df[(df['state'] == 'CA') & (df['type'] == 'villa')]

# Find the row with the minimum price
cheapest_villa = ca_villas.loc[ca_villas['price'].idxmin()]

# Retrieve the city, street, and price of the cheapest villa
cheapest_city = cheapest_villa['city']
cheapest_street = cheapest_villa['street']
cheapest_price = cheapest_villa['price']

# Print the result
print(f"The cheapest villa in CA is located in {cheapest_city}, {cheapest_street} and has a price of ${cheapest_price}")



```

## List top 5 residency details which lie in the budget of 60000-120000, an area more than 1450, min bedroom 3 and, min bathroom 2.



```python
# filter the data based on criteria
filtered_data = df[
    (df['price'] >= 60000) &
    (df['price'] <= 120000) &
    (df['sq__ft'] > 1450) &
    (df['beds'] >= 3) &
    (df['baths'] >= 2)
]
top_5_residencies=filtered_data.sort_values(by='price',ascending=True).head(5)
top_5_residencies
```
