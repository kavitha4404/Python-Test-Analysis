```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('auto_data')
df
```

```python
df.info()
```

```python
# converting a string value to an equivalent number value

# Convert the data type of 'horsepower' to int
df['horsepower'] =pd.to_numeric(df['horsepower'],errors='coerce')
```

```python
# display the statistical data
df.describe()
```

```python
# Display the top 5 records
top5_records=df.head(5)
print('The Top5 Records:')
top5_records
```

```python
#  display the bottom 5 records
bottom5 = df.tail(5)
print('The Bottom5 Records:')
bottom5
```

```python
# Displaying the Correlation between variables
sns.heatmap(df.corr(),annot=True)
```

```python
# Grouping the data by a categorical variable('origin') and calculating mean
grouped_data = df.groupby('origin').mean()
grouped_data
```

```python
# Barplot
sns.barplot(x=grouped_data.index, y='mpg', data=grouped_data)
plt.title('Mean MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('Mean MPG')
plt.show()


```

```python
# barplot
sns.barplot(x=grouped_data.index, y='displacement', data=grouped_data)
plt.title('Mean displacement by Origin')
plt.xlabel('Origin')
plt.ylabel('Mean diplacement')
plt.show()
```

```python
# line plot
sns.lineplot(x=grouped_data.index, y='weight', data=grouped_data)
plt.title('Mean Weight by Origin')
plt.xlabel('Origin')
plt.ylabel('Mean Weight')
plt.show()
```

```python
# boxplot
sns.boxplot(x=grouped_data.index, y='acceleration', data=grouped_data)
plt.title('Mean Weight by acceleration')
plt.xlabel('Origin')
plt.ylabel('Mean acceleration')
plt.show()
```

```python
# The pair plot for mpg, weight,orign
sns.pairplot(df[['mpg','weight','origin']],hue='origin',size=4)
```


## Identify how many types of  cylinder vehicles

```python
# Group by the 'cylinders' column and count the number of unique values
by_cylinders = df.groupby('cylinders')['name'].nunique()
num_cylinder_types = len(by_cylinders)
print("Number of unique cylinder types:", num_cylinder_types)
print("Cylinder types:", by_cylinders)
```

```python
# Count the number of vehicles for each model year
vehicle_counts = df["model_year"].value_counts().sort_index()

# Display the results
df1 = pd.DataFrame({'model_year':vehicle_counts.index,'vehicle_count':vehicle_counts.values})
df1
```
