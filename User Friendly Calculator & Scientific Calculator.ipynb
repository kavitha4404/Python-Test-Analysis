### Build an User Friendly Calculator

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
print("1.Addition")
print("2.Subtraction")
print("3.Multilication")
print("4.Division")
print("Please choose your option from 1/2/3/4")
choice = int(input("Enter your option:"))
num1 = float(int(input("Enter the First Number")))
num2 = float(int(input("Enter the Second Number")))
if choice == 1:
    n = add(num1,num2)
    print("The addition value is:",n)
elif choice == 2:
    n = sub(num1,num2)
    print("The subtraction value is:",n)
elif choice == 3:
    n = mul(num1,num2)
    print("The multiplication value is:",n)
elif choice == 4:
    n = div(num1,num2)
    print("The division value is:",n)
else:
    print(" please enter a valid number")
```

### Build a Scientific Calculator

```python
import math
```

```python
while True:
    print("\n Choose the math Operation's")
    print("0.Addition")
    print("1.Subtraction")
    print("2.Multilication")
    print("3.Division")
    print("4.Module")
    print("5.Raising Of Power")
    print("6.Square Root")
    print("7.Logarithms")
    print("8.Sine")
    print("9.Cosine")
    print("10.Tangent")
    oper = input("\n Your Option from Menu")
    # addition
    if oper == "0":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(val1+val2) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break



    # Subtraction
    elif oper == "1":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(val1-val2) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Multiplication
    elif oper == "2":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(val1*val2) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Division
    elif oper == "3":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(val1/val2) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break

    # Modulus
    elif oper == "4":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(val1%val2) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break

    # Raising Of Power
    elif oper == "5":
        val1 = float(int(input("\n Enter First Value")))
        val2 = float(int(input("\n Enter Second Value")))
        print("\n The result is:" + str(math.pow(val1,val2)) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Square Root
    elif oper == "6":
        val1 = float(int(input("\n Enter First Value")))
        print("\n The result is:" + str(math.sqrt(val1)) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Logarithms
    elif oper == "7":
        val1 = float(int(input("\n Enter First Value")))
        print("\n The result is:" + str(math.log(val1)) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Sine
    elif oper == "8":
        val1 = float(int(input("\n Enter First Value")))
        print("\n The result is:" + str(math.sin(math.radians(val1))) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Cosine
    elif oper == "9":
        val1 = float(int(input("\n Enter First Value")))
        print("\n The result is:" + str(math.cos(math.radians(val1))) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break
        
    # Tangent
    elif oper == "10":
        val1 = float(int(input("\n Enter First Value")))
        print("\n The result is:" + str(math.tan(math.radians(val1))) + "\n")
        back = input("\n Go back to the main menu ?(Y/N)")
        if back == 'Y':
            continue
        else:
            break

    # Handling Invalid Data
    else:
        print("\n Invalid Option")
        continue
    
```
