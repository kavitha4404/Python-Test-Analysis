### 1. Find out the total number of cards approved by States. 

```python
# Group the data by 'State' and sum the 'Aadhaar generated' column for each state
cards_approved_by_states = df.groupby('State')['Aadhaar generated'].sum()
df1 = pd.DataFrame(cards_approved_by_states)
df1
```

### 2. Find out the total number of cards Enrolment rejected by states. 


```python
# Group the data by 'State' and sum the 'Enrolment rejected' column for each state
cards_enrolment_rej_states = df.groupby('State')["Enrolment Rejected"].sum()
df1 = pd.DataFrame(cards_enrolment_rej_states)
df1
```

### 3. Find out the total number of cards approved by Districts.

```python
# Group the data by 'District' and sum the 'Aadhaar generated' column for each District
cards_approved_by_district = df.groupby('District')['Aadhaar generated'].sum()
df1 = pd.DataFrame(cards_approved_by_district)
df1
```

### 4. Find out the total number of cards rejected by Districts.


```python
# Group the data by 'State' and sum the 'Enrolment rejected' column for each state
cards_enrolment_rej_District = df.groupby('District')["Enrolment Rejected"].sum()
df1 = pd.DataFrame(cards_enrolment_rej_District)
df1
```

### 5.Total Count Gender Based 

```python
# Group the data by 'Gender' and count the number of occurrences for each gender
gender_count = df['Gender'].value_counts()
df1 = pd.DataFrame({"gender":gender_count.index,"count":gender_count.values})
df1
```

### 6.Count Records age <25, age between 25 &35, age>45 

```python
# Defining the age categories
Age_Categories = {'age<25':(df['Age'] < 25).sum(),
                 'age(25-35)':((df['Age'] >= 25) & (df['Age'] <= 35)).sum(),
                 'age>45':(df['Age'] > 45).sum()}
# print the count of records for each age category
print(Age_Categories)
```
